
In Python, the max() function is used to find the maximum element from a sequence or multiple arguments.

```python
import random
```

```python
random_numbers = list()
for i in range(0,10000):
  blank_string = str()

  random_number = random.randint(0,35)
  blank_string = blank_string + mapping_dict[random_number]

  random_number = random.randint(0,35)
  blank_string = blank_string + mapping_dict[random_number]


  for i in range(0,8):
    coin_toss = random.randint(0,1)
    if coin_toss == 1:
      random_number = random.randint(0,35)
      blank_string = blank_string + mapping_dict[random_number]

    else:
      break
  random_numbers.append(blank_string)
```

A loop will run 1000 times and two random numbers will be generated, and will get stored in the blank_string.

then other loop will run 8 times, then we will toss a coin to get a random value 0 or 1

then we will repeat the same process

```python
random_numbers
```

```python
base_frequency = dict()


for blank_string in random_numbers:
    highest_digit = max(blank_string)
    highest_base = reverse_mapping_dict[highest_digit] + 1

    if highest_base in base_frequency.keys():
      base_frequency[highest_base] += 1
    else:
      base_frequency[highest_base] = 1


    print("highest base of {} is {}".format(blank_string, highest_base))
```

A new empty dictionary named base_frequency is initialized. Now, we iterate through each blank_string in the random_numbers list. For each blank_string, we find the highest base using the max() function, and if this highest base is already a key in the base_frequency dictionary, we increment its corresponding value by 1. If the highest base is not present in the dictionary, we add it as a new key with the initial value of 1.

```python
print(base_frequency)
```
